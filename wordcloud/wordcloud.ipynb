```python
fi = open('routeinfo.csv', 'r', encoding = 'UTF-8')
routeinfo = {}
for line in fi.readlines():
    if ',' not in line:
        line = line[:-1]
        key = line
    else:
        value = line.split(',')
        routeinfo[key] = value
fi.close()
fi = open('addressinfo.csv', 'r', encoding = 'UTF-8')
addressinfo = {}
for line in fi.readlines():
    temp = line.split(',')
    key = temp[0]
    addressinfo[key] = line
fi.close()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter; t = Twitter()
```

```python
ko_con_text = open('wordcloud.txt', encoding='UTF-8').read()
tokens_ko = t.nouns(ko_con_text)
ko = nltk.Text(tokens_ko, name='test')
data = ko.vocab().most_common(500)
tmp_data = dict(data)
```

```python
wc = WordCloud(font_path = "c:/Windows/Fonts/malgun.ttf", 
               relative_scaling = 0.2, 
               background_color = 'white', mask = korea_coloring)
wc = wc.generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(wc.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
fig = plt.gcf()
plt.show()
fig.savefig('구별wordclouding.png')
```
