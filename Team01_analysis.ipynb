## 위도, 경도 정보와 지구 반지름, 각도 등의 정보를 이용하여 정류소간 거리 계산

```python
from math import sin, cos, sqrt, atan2, radians
import csv

f = open('data/routeinfo.csv', 'r', encoding = 'UTF-8')
ff = open('data/gpsinfo.csv', 'r', encoding = 'UTF-8')
gpsinfo = {}
r = ff.read().splitlines()
for line in r:
    t = line.split(',')
    key = t[0]
    value = [t[1], t[2]]
    gpsinfo[key] = value
del r
ff.close()

route_id = []
i = 1
for line in f.readlines():
    if ',' not in line:
        title = 'route_distance/' + line[:-1] + '.csv'
        fi = open(title, 'w', encoding = 'UTF-8', newline = '')
        route_id.append(line[:-1])
    else:
        temp_list = line.split(',')
        l = 0
        distance_result = []
        while l < len(temp_list) - 2:
            a = temp_list[l]
            b = temp_list[l + 1]
            lat1 = radians(float(gpsinfo[a][1]))
            lon1 = radians(float(gpsinfo[a][0]))

            lat2 = radians(float(gpsinfo[b][1]))
            lon2 = radians(float(gpsinfo[b][0]))
            
            ##위도, 경도로 계산하는 부분            
            R = 6371.0
            dlon = lon2 - lon1
            dlat = lat2 - lat1
            cal = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
            cal2 = 2* atan2(sqrt(cal), sqrt(1 - cal))
            distance = R * cal2
            distance_result.append(distance)
            l = l + 1
        wr = csv.writer(fi)
        wr.writerow(distance_result)
        fi.close()
    i = i + 1
f.close()

with open('data/route_id.json', 'w') as f:
    json.dump(route_id, f)
```
